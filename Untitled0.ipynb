```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Titanic.csv")
df.head(5)
```

```python
df=df.dropna(subset=["Survived","Age"])

```

```python
#feature are the column that help to predict the target
# Example: Using 'Pclass', 'Age', 'Fare', and 'Sex' as features
# The 'target' is the column we want to predict, which is 'Survived'

# Filter data for two groups: survived (1) and not survived (0)

survived_age= df[df["Survived"]==1]["Age"]
Not_survived =df[df["Survived"]==0]["Age"]


#simulation parameter
alpha=0.05
num_simulations = 1500
sample_size = 25

#initialize counters for errors
type1_count=0
type2_count=0

# Type 1 Error Simulation: Compare survived vs. survived (identical distribution)

for _ in range(num_simulations):
    group1=np.random.choice(survived_age, sample_size, replace=True )
    group2=np.random.choice(survived_age, sample_size, replace=True)

    t_stats,p_value = ttest_ind(group1,group2)
    if p_value < alpha:
        type1_count+=1

# Type 2 Error Simulation: Compare survived vs. not survived (different distributions)
for _ in range(num_simulations):
    group1=np.random.choice(survived_age, sample_size, replace=True)
    grouup2=np.random.choice(Not_survived, sample_size, replace=True)

    t_stats, p_value= ttest_ind(group1,group2)
    if p_value >= alpha:
        type2_count+=1


#calculate error rate

type1_rate=type1_count / num_simulations
type2_rate=type2_count / num_simulations

#print the result


print("type1_ error rate(False Positive)", type1_rate)
print("type2_ error rate(False Negative)", type2_rate)
```

```python

# Visualization 1: Histograms for Age of Survivors and Non-Survivors

plt.figure(figsize=(10,5))
plt.hist(survived_age, bins=10, alpha=0.7, label="Survived",color='blue')
plt.hist(Not_survived, bins=10, alpha=0.7, label="Not Survived", color='green')
plt.title("Distribution of Age for Survivors and Non-Survivors")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.legend()
plt.show()
```

```python
# Visualization 2: Type 1 and Type 2 Error Rates as a Bar Chart
error_rates = [type1_rate, type2_rate]
error_labels = ['Type 1 Error Rate', 'Type 2 Error Rate']
plt.figure(figsize=(8, 6))
plt.bar(error_labels, error_rates, color=['red', 'orange'])
plt.title("Type 1 and Type 2 Error Rates")
plt.ylabel("Error Rate")
plt.show()
```
